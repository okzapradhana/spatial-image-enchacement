```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Image Negative
def convertRGBToGrayLevel(img):
    Red = img[:,:,0]
    Green = img[:,:,1]
    Blue = img[:,:,2]
    grayLevel = 0.21 * Red + 0.71 * Green + 0.07 * Blue
    arrayGrayLevel = np.uint8(grayLevel*255)
    return (arrayGrayLevel)

def negative(img):
    L = np.max(img)
    neg = L - img
    return neg

img = plt.imread("./Lenna.png")
gray = convertRGBToGrayLevel(img)
negative = negative(gray)

plt.imshow(img)
plt.title("Original Image")
plt.show()

plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.title("gray")
plt.show()


plt.imshow(negative, cmap = plt.get_cmap('gray'))
plt.title("negative image")
plt.show()
```

```python
#Log Transformations
def logTransformations(img, c):
    img = img / 255
    S = c * np.log(1 + img)
    S = np.uint8(S*255)
    return S

img = plt.imread("Lenna.png")
gray = convertRGBToGrayLevel(img)
imgLog = logTransformations(gray, 1)

plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.title("Ori image")
plt.show()

plt.imshow(imgLog, cmap = plt.get_cmap('gray'))
plt.title("Log transformation")
plt.show()
```

```python
#Power Law Transformations
def powerLawTransformation(img, c, y):
    img = img/255
    hasil = c * (np.power(img,y))
    return hasil

imgLaw = plt.imread("TestImageLaw.png")
gray = convertRGBToGrayLevel(imgLaw)
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.title("Original Image")
plt.show()

imgPowLaw = powerLawTransformation(gray, 1, 3)
plt.imshow(imgPowLaw, cmap = plt.get_cmap('gray'))
plt.title("Gamma = 3")
plt.show()
```

```python
#Piecewise Linear Transformations

imgPiecewise = plt.imread("./Lenna.png")
gray = convertRGBToGrayLevel(imgPiecewise)

print(gray)

maxGrayLevel = np.max(gray)
print(maxGrayLevel)

plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.title("Original Image")
plt.show()

x = [0, 127 ,140 , maxGrayLevel]
y = [0, 123, 150, maxGrayLevel]

plt.plot(x, y)
plt.title("Grafik untuk Piecewise Linear")
plt.axis([0, maxGrayLevel, 0, maxGrayLevel])
plt.show()

gray[gray < x[1]] = 0
gray[(gray >= x[1]) & (gray <= x[3]) ] = (255/2)
gray[gray > x[3]] = 255

plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.title("Setelah dilakukan Piecewise Linear Transformations")
plt.show()
```

```python
def bitPlane(img,plane):
    plane = 7 - plane
    [x,y] = np.shape(img)
    a = np.zeros((x,y))
    l = list()
    
    for i in img:
        l.append(list(map(lambda x: 0 if np.unpackbits(x)[plane] == 0 else 255, i)))
    
    a = np.array(l);
    return a

def rgb2Gray(img):
    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]
    gray = 0.21 * r + 0.71 * g + 0.07 *b
    return np.uint8(gray*255);

for i in range (0,8):
    plt.imshow(bitPlane(rgb2Gray(img),i), cmap="gray")
    plt.title("plane - %i"%i)
    plt.show()
```

```python
#Substraction
img1 = plt.imread("dataset1.jpeg")
plt.imshow(img1)
plt.title("First Image")
plt.show()

img2 = plt.imread("dataset2.jpeg")
plt.imshow(img2)
plt.title("Second Image")
plt.show()

imgSubstraction = np.subtract(img1, img2)
plt.imshow(imgSubstraction, cmap = plt.get_cmap('gray'))
plt.title("Setelah dilakukan substraction")
plt.show()
```

```python
#Logical operations

#OR
imageMask = plt.imread("logicaloperations.png")
imgMask = convertRGBToGrayLevel(imageMask)
plt.imshow(imgMask, cmap = plt.get_cmap('gray'))
plt.title("OR image Mask")
plt.show()

oriImage = plt.imread("Lenna.png")
gray = convertRGBToGrayLevel(oriImage)

plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.title("Gray Image")
plt.show()

ORLogicalImage = np.bitwise_or(imgMask, gray)

plt.imshow(ORLogicalImage, cmap = plt.get_cmap('gray'))
plt.title("Setelah dilakukan OR Logical Image Operations")
plt.show()
```

```python
#AND
imageMask = plt.imread("logicaloperations.png")
imgMask = convertRGBToGrayLevel(imageMask)
plt.imshow(imgMask, cmap = plt.get_cmap('gray'))
plt.title("AND image Mask")
plt.show()

oriImage = plt.imread("Lenna.png")
gray = convertRGBToGrayLevel(oriImage)

plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.title("Gray Image")
plt.show()

ANDLogicalImage = np.bitwise_and(imgMask, gray)

plt.imshow(ANDLogicalImage, cmap = plt.get_cmap('gray'))
plt.title("Setelah dilakukan OR Logical Image Operations")
plt.show()
```
